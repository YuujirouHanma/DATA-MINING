```python
# Binning untuk Num_Killed
bins_killed = [0, 10, 20, 30, 40, 50, np.inf]  # Sesuaikan rentang bin
labels_killed = ['0-10', '11-20', '21-30', '31-40', '41-50', '50+']  # Label bin
data['Num_Killed_Binned'] = pd.cut(data['Num_Killed'], bins=bins_killed, labels=labels_killed)

# Binning untuk Total_Crashes
bins_crashes = [0, 50, 100, 150, 200, np.inf]  # Sesuaikan rentang bin
labels_crashes = ['0-50', '51-100', '101-150', '151-200', '200+']  # Label bin
data['Total_Crashes_Binned'] = pd.cut(data['Total_Crashes'], bins=bins_crashes, labels=labels_crashes)

# Binning untuk Num_Injured
bins_injured = [0, 50, 100, 150, 200, 300, np.inf]  # Sesuaikan rentang bin
labels_injured = ['0-50', '51-100', '101-150', '151-200', '201-300', '300+']  # Label bin
data['Num_Injured_Binned'] = pd.cut(data['Num_Injured'], bins=bins_injured, labels=labels_injured)

# Binning untuk Total_Vehicles_Involved
bins_vehicles = [0, 50, 100, 150, 200, np.inf]  # Sesuaikan rentang bin
labels_vehicles = ['0-50', '51-100', '101-150', '151-200', '200+']  # Label bin
data['Total_Vehicles_Involved_Binned'] = pd.cut(data['Total_Vehicles_Involved'], bins=bins_vehicles, labels=labels_vehicles)

# Membuat plot countplot untuk semua variabel yang telah dibinned
fig, axes = plt.subplots(2, 2, figsize=(18, 18))
fig.suptitle('Bar plot for binned variables in the dataset')

# Countplot untuk Num_Killed_Binned
sns.countplot(ax=axes[0, 0], x='Num_Killed_Binned', data=data, color='red')
axes[0, 0].set_title('Num Killed')

# Countplot untuk Total_Crashes_Binned
sns.countplot(ax=axes[0, 1], x='Total_Crashes_Binned', data=data, color='blue')
axes[0, 1].set_title('Total Crashes')

# Countplot untuk Num_Injured_Binned
sns.countplot(ax=axes[1, 0], x='Num_Injured_Binned', data=data, color='green')
axes[1, 0].set_title('Num Injured')

# Countplot untuk Total_Vehicles_Binned
sns.countplot(ax=axes[1, 1], x='Total_Vehicles_Involved_Binned', data=data, color='purple')
axes[1, 1].set_title('Total Vehicles Involved')

# Mengatur rotasi dan layout agar lebih rapi
for ax in axes.flat:
    plt.setp(ax.get_xticklabels(), rotation=45)

plt.tight_layout()
plt.show()
```

```python

def log_transform(data,col):
    for colname in col:
        if (data[colname] == 1.0).all():
            data[colname + '_log'] = np.log(data[colname]+1)
        else:
            data[colname + '_log'] = np.log(data[colname])
    data.info()
```

```python
log_transform(data,['Total_Crashes','Num_Injured', 'Num_Killed', 'Total_Vehicles_Involved'])
```

```python
# Binning untuk Num_Killed
bins_killed = [0, 10, 20, 30, 40, 50, np.inf]  # Sesuaikan rentang bin
labels_killed = ['0-10', '11-20', '21-30', '31-40', '41-50', '50+']  # Label bin
data['Num_Killed_Binned'] = pd.cut(data['Num_Killed_log'], bins=bins_killed, labels=labels_killed)

# Binning untuk Total_Crashes
bins_crashes = [0, 50, 100, 150, 200, np.inf]  # Sesuaikan rentang bin
labels_crashes = ['0-50', '51-100', '101-150', '151-200', '200+']  # Label bin
data['Total_Crashes_Binned'] = pd.cut(data['Total_Crashes_log'], bins=bins_crashes, labels=labels_crashes)

# Binning untuk Num_Injured
bins_injured = [0, 50, 100, 150, 200, 300, np.inf]  # Sesuaikan rentang bin
labels_injured = ['0-50', '51-100', '101-150', '151-200', '201-300', '300+']  # Label bin
data['Num_Injured_Binned'] = pd.cut(data['Num_Injured_log'], bins=bins_injured, labels=labels_injured)

# Binning untuk Total_Vehicles_Involved
bins_vehicles = [0, 50, 100, 150, 200, np.inf]  # Sesuaikan rentang bin
labels_vehicles = ['0-50', '51-100', '101-150', '151-200', '200+']  # Label bin
data['Total_Vehicles_Involved_Binned'] = pd.cut(data['Total_Vehicles_Involved_log'], bins=bins_vehicles, labels=labels_vehicles)

# Membuat plot countplot untuk semua variabel yang telah dibinned
fig, axes = plt.subplots(2, 2, figsize=(18, 18))
fig.suptitle('Bar plot for binned variables in the dataset')

# Countplot untuk Num_Killed_Binned
sns.countplot(ax=axes[0, 0], x='Num_Killed_Binned', data=data, color='red')
axes[0, 0].set_title('Num Killed')

# Countplot untuk Total_Crashes_Binned
sns.countplot(ax=axes[0, 1], x='Total_Crashes_Binned', data=data, color='blue')
axes[0, 1].set_title('Total Crashes')

# Countplot untuk Num_Injured_Binned
sns.countplot(ax=axes[1, 0], x='Num_Injured_Binned', data=data, color='green')
axes[1, 0].set_title('Num Injured')

# Countplot untuk Total_Vehicles_Binned
sns.countplot(ax=axes[1, 1], x='Total_Vehicles_Involved_Binned', data=data, color='purple')
axes[1, 1].set_title('Total Vehicles Involved')

# Mengatur rotasi dan layout agar lebih rapi
for ax in axes.flat:
    plt.setp(ax.get_xticklabels(), rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Menghitung korelasi menggunakan metode Pearson
correlation_matrix = data[['Total_Crashes_log', 'Num_Injured_log', 'Num_Killed_log', 
                           'Total_Vehicles_Involved_log', 'SPV', 'DAD', 'PWR', 
                           'FTQ']].corr()

# Menampilkan matriks korelasi
print(correlation_matrix)

# Visualisasi menggunakan heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Variables')
plt.show()
```
