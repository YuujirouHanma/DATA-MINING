HANDLING OUTLIERS

```python
def check_outlier(data, var):
    Q1 = data[var].quantile(0.25)
    Q3 = data[var].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[(data[var] < lower_bound) | (data[var] > upper_bound)]
    return outliers

outlier_time_spend = check_outlier(data=data, var="time_spend_company")
outlier_time_spend
```

BAR PLOT UNTUK VARIABEL KATEGORIKAL

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 8))
fig.suptitle('Bar plot for categorical variables: dept and salary')

sns.countplot(ax=axes[0], x='dept', data=data, color='red',
              order=data['dept'].value_counts().index)
axes[0].set_title('Distribution of Dept')
axes[0].tick_params(labelrotation=45)

sns.countplot(ax=axes[1], x='salary', data=data, color='red',
              order=data['salary'].value_counts().index)
axes[1].set_title('Distribution of Salary')
axes[1].tick_params(labelrotation=45)

plt.tight_layout(rect=[0, 0, 1, 0.95]) 
plt.show()
```

BAR PLOT UNTUK VARIABEL NUMERIKAL

```python
numerical_columns = ['satisfaction_level', 'last_evaluation', 'number_project', 
                     'average_montly_hours', 'time_spend_company', 
                     'Work_accident', 'promotion_last_5years']

# Buat subplots untuk numerik variabel
fig, axes = plt.subplots(4, 2, figsize=(18, 20))
fig.suptitle('Histogram for numerical variables', fontsize=16)

for i, column in enumerate(numerical_columns):
    row = i // 2  # Determine the row
    col = i % 2   # Determine the column
    sns.histplot(ax=axes[row, col], data=data, x=column, bins=20, color='blue', kde=True)
    axes[row, col].set_title(f'Distribution of {column}')
    axes[row, col].set_xlabel(column)
    axes[row, col].set_ylabel('Frequency')

# Hide the last empty subplot
axes[3, 1].axis('off')

plt.tight_layout(rect=[0, 0, 1, 0.95]) 
plt.show()
```

MULTIVARIATE ANALYSIS

```python
label_data = ['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 'time_spend_company', 'Work_accident','promotion_last_5years']
scaler = StandardScaler()
data_scaled = data.copy()
data_scaled[label_data] = scaler.fit_transform(data[label_data])

correlation_matrix = data_scaled[label_data].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Multivariate Analysis: Correlation Matrix')
plt.show()
```
